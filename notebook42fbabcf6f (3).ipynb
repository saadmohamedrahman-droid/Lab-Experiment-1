```python
import matplotlib.pyplot as plt

# Data for left plot (Training Accuracy)
epochs = [1, 2, 3, 4, 5]
accuracy = [0.55, 0.72, 0.81, 0.89, 0.94]

# Data for right plot (Inference Speed)
hardware = ['CPU', 'GPU', 'TPU']
fps = [15, 120, 350]

# Create a figure
plt.figure(figsize=(10, 4))

# Left Plot: Line Plot for Training Accuracy 
plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, marker='o', color='blue')
plt.title("Model Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (0.0-1.0)")
plt.grid(True)

# Bar Chart for Hardware Performance
plt.subplot(1, 2, 2)
plt.bar(hardware, fps,
        color=['tomato', 'royalblue', 'mediumpurple'],
        edgecolor='black')

plt.title('Inference Speed Comparison', fontsize=12, fontweight='bold')
plt.xlabel('Hardware Setup')
plt.ylabel('Frames Per Second (FPS)')
plt.grid(axis='y', linestyle=':', alpha=0.6)
# Adjust layout
plt.tight_layout()

# Layout formatting and display
plt.tight_layout()
plt.show()
```
